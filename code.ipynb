```python
class Task:
    def __init__(self, title, description):
        self.title = title
        self.description = description
        self.completed = False

class Planner:
    def __init__(self):
        self.tasks = []

    def add_task(self, title, description):
        task = Task(title, description)
        self.tasks.append(task)
        print("Task '{}' added successfully.".format(title))

    def view_tasks(self):
        if not self.tasks:
            print("No tasks available.")
            return
        print("Tasks:")
        for i, task in enumerate(self.tasks, start=1):
            print("{}. {} - {}".format(i, task.title, "Completed" if task.completed else "Pending"))

    def mark_task_completed(self, task_number):
        index = task_number - 1
        if 0 <= index < len(self.tasks):
            self.tasks[index].completed = True
            print("Task '{}' marked as completed.".format(self.tasks[index].title))
        else:
            print("Invalid task number.")

    def delete_task(self, task_number):
        index = task_number - 1
        if 0 <= index < len(self.tasks):
            del self.tasks[index]
            print("Task deleted successfully.")
        else:
            print("Invalid task number.")

# Example usage:
planner = Planner()

while True:
    print("\nPlanner Menu:")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        title = input("Enter task title: ")
        description = input("Enter task description: ")
        planner.add_task(title, description)
    elif choice == '2':
        planner.view_tasks()
    elif choice == '3':
        task_number = int(input("Enter task number to mark as completed: "))
        planner.mark_task_completed(task_number)
    elif choice == '4':
        task_number = int(input("Enter task number to delete: "))
        planner.delete_task(task_number)
    elif choice == '5':
        print("Exiting planner. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
